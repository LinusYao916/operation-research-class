```python
import gurobipy as gp

# 店铺和收入
stores = list(range(1, 11))
revenue = {
    1:127, 2:83, 3:165, 4:96, 5:112, 6:88,
    7:135, 8:141, 9:117, 10:94
}

# 圈内分组（最多只能选2个）
regions = [
    [1, 3],
    [1, 2, 4],
    [4, 5, 6],
    [6, 7, 8],
    [6, 9],
    [8, 10],
    [9, 10]
]

# 模型
m = gp.Model('maximize_revenue')

# 决策变量：是否选这个店
x = m.addVars(stores, vtype=gp.GRB.BINARY, name="x")

# 目标：最大总收入
m.setObjective(gp.quicksum(revenue[i] * x[i] for i in stores), gp.GRB.MAXIMIZE)

# 每个圈内最多只能选两个
for group in regions:
    m.addConstr(gp.quicksum(x[i] for i in group) <= 1)

# 求解
m.optimize()

# 输出结果
selected = [i for i in stores if x[i].x > 0.9]
print("选择的店铺编号:", selected)
print("总收益（$1000s）:", sum(revenue[i] for i in selected))

```
